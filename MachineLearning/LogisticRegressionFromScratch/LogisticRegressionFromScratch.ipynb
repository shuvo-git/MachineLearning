```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import seaborn as sns
import math
import pandas as pd
import tkinter as tk
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x =[]
y = []
for i in range(-100,100):
  x.append(i)
  y.append(sigmoid(i))
```

```python
fig = plt.figure(figsize=(10,5))
graph = fig.add_subplot(111)
#for i in range(len(x)):

plt.plot(x, y, color='green',marker=',')
plt.show()
```

```python

data = np.array([x,y])
df = pd.DataFrame(data.T,columns=['X','Sigmoid Value']) 
g = sns.relplot(x="X", y="Sigmoid Value",height=5,aspect=2,data=df)
```
