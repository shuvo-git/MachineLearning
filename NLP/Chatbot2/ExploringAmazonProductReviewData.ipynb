```
import pandas as pd
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from bs4 import BeautifulSoup
import string
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
```

```
df.shape
df['number_of_reviews']= df['number_of_reviews'].fillna(0)
df['number_of_reviews'] = df['number_of_reviews'].apply(lambda x:  re.sub(',','',str(x))  )
df['number_of_reviews'] = df['number_of_reviews'].apply(int)
```

```
#df.isnull().mean().sort_values(ascending=False)
reviews_df = df.drop(['customer_questions_and_answers','sellers','items_customers_buy_after_viewing_this_item','number_available_in_stock','price','customers_who_bought_this_item_also_bought','number_of_answered_questions'],axis=1)
reviews_df = reviews_df.dropna()
reviews_df.isnull().mean().sort_values(ascending=False)
reviews_df.shape
```

```
reviews = reviews_df['customer_reviews'].str.split("//",n=4,expand=True)
reviews[3] = reviews[3].str.split("\n",n=3,expand=True)[2]

reviews_df['review_title'] = reviews[0]
reviews_df['rating'] = reviews[1]
reviews_df['review_date'] = reviews[2]
reviews_df['customer_name'] = reviews[3]
reviews_df['review'] = reviews[4]
review_df = reviews_df.drop(columns=['customer_reviews'])
review_df.shape
```

```
tokenizer = RegexpTokenizer(r'\w+')
review_df['review'] = review_df['review'].apply(lambda x: tokenizer.tokenize(x.lower() ) )
review_df['review'].head(20)
```

```
def removeStopWords(words):
  rmv_stp_words = [w for w in words if w not in stopwords.words('english')]
  return rmv_stp_words

len(stopwords.words('english'))
review_df['review'] = review_df['review'].apply(lambda x: removeStopWords(x))
review_df['review']
```

```
# Word Lemmetizing

print(review_df['review'])
lemmatizer = WordNetLemmatizer()
def wl(text):
  lem_text = [ lemmatizer.lemmatize(i) for i in text ]
  return lem_text

review_df_lem = review_df['review'].apply(lambda x:  wl(x) ) 
print(review_df_lem)
```

```
#Word Stemming

stemmer = PorterStemmer()

def wordStemmer(text):
  stem_text = [ stemmer.stem(i) for i in text ]
  return stem_text

review_df_stem = review_df['review'].apply(lambda x: wordStemmer(x) )
review_df_lem
```
